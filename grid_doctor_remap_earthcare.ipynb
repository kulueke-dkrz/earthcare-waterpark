```python
import earthcarekit as eck
import grid_doctor as gd

eck.ecdownload("ACM_CAP_2B", start_time="2025-08-10T00:00:00", end_time="2025-08-10T06:00:00")
products = eck.search_product(file_type="ACM_CAP_2B")

ds = eck.read_product(products.filepath[0])

finest = gd.bin_to_healpix(
    ds[["ice_water_path", "longitude", "latitude"]],
    level=11,
    agg="mean",
    lat_name="latitude",
    lon_name="longitude",
)
pyramid = {11: finest}
for level in range(10, -1, -1):
    pyramid[level] = gd.coarsen_healpix(pyramid[level + 1], level)

gd.save_pyramid(pyramid, "/work/ks1387/gw/data/obs/earthcare/2024-08-10/")
```
