```python
import earthcarekit as eck
import grid_doctor as gd

eck.ecdownload("ACM_CAP_2B", start_time="2025-08-09T00:00:00", end_time="2025-08-09T06:00:00")
products = eck.search_product(file_type="ACM_CAP_2B")

ds = eck.read_product(products.filepath[0])

finest = gd.bin_to_healpix(
    ds,                                   # per-sample latitude/longitude
    level=11,                             # coarser than the sample spacing!
    agg="mean",
    fill_values={"cloud_type": 255},
    with_counts=True,                     # auditable coverage
)
pyramid = {11: finest}
for level in range(10, -1, -1):
    pyramid[level] = gd.coarsen_healpix(pyramid[level + 1], level)
gd.save_pyramid(pyramid, "/work/ks1387/gw/data/obs/earthcare/2024-08-09/")
```
